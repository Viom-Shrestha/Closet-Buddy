## Task 1.1: The Pixel Grid – Understanding Images as a Matrix



Machines do not "see" faces, trees, or objects.  
They see a grid of pixel intensity values ranging from **0 to 255**.


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Step 1: Load a JPEG Image



```python
img = cv2.imread('sample.jpeg')

print("Image Shape:", img.shape)
print("Data Type:", img.dtype)
```

### What does the shape mean?

If the output is something like:

(1200, 1200)

It means:
- 1200 rows (height)
- 1200 columns (width)

There is NO third dimension because this is a grayscale image.


```python
plt.imshow(img, cmap='gray')
plt.title("Image")
plt.axis("off")
plt.show()
```

## Step 2: Inspect a 10×10 Pixel Slice

Now we will print a small section of the image matrix.

This proves that the image is simply a grid of numbers.


```python
# Select a 10x10 region
row_start, col_start = 100, 100
pixel_slice = img[row_start:row_start+10, col_start:col_start+10]

print("10x10 Pixel Slice:\n")
print(pixel_slice)
```

## Step 4: Visualize the Selected Region

Let’s draw a red square around the selected area so we know
where those numbers came from.


```python
img_color = img

# Draw rectangle (red box)
cv2.rectangle(img_color,
              (col_start, row_start),
              (col_start+100, row_start+100),
              (0,0,255), 2)

# Display
plt.figure(figsize=(6,6))
plt.imshow(cv2.cvtColor(img_color, cv2.COLOR_BGR2RGB))
plt.title("Selected 10x10 Region Highlighted")
plt.axis("off")
plt.show()
```
