## Task 1.2: The Color Channel Mystery

A color image is stored as a **3D NumPy array**:

(Height, Width, 3)

The third dimension represents color channels.


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

## Step 1: Load a Color Image

We load the image normally using:

cv2.imread(path)

By default, OpenCV loads images in **BGR format**.


```python
img = cv2.imread("sample.jpeg")

if img is None:
    print("Image not found. Check file path.")
else:
    print("Image loaded successfully!")

print("Image Shape:", img.shape)
print("Data Type:", img.dtype)
```

```python
# Convert BGR to RGB for correct display
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(6,6))
plt.imshow(img_rgb)
plt.title("Original Color Image")
plt.axis("off")
plt.show()
```

## Step 1: Split the Color Channels

We separate the image into:

- Blue channel
- Green channel
- Red channel

Each channel is a 2D matrix.


```python
b, g, r = cv2.split(img)

print("Blue Channel Shape:", b.shape)
print("Green Channel Shape:", g.shape)
print("Red Channel Shape:", r.shape)
```

```python
image_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
image_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

```python
image_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
image_lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)
```

```python
plt.figure(figsize=(12, 8))

# 1 Original
plt.subplot(2, 3, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Original (BGR)")
plt.axis("off")

# 2 RGB
plt.subplot(2, 3, 2)
plt.imshow(image_rgb)
plt.title("RGB")
plt.axis("off")

# 3 HSV
plt.subplot(2, 3, 3)
plt.imshow(image_hsv)
plt.title("HSV")
plt.axis("off")

# 4 Grayscale
plt.subplot(2, 3, 4)
plt.imshow(image_gray, cmap='gray')
plt.title("Grayscale")
plt.axis("off")

# 5 LAB
plt.subplot(2, 3, 5)
plt.imshow(image_lab)
plt.title("LAB")
plt.axis("off")

plt.tight_layout()
plt.show()
```

Now we remove all red information from the image.

What do you think will happen?


```python
# Create a copy
img_no_red = img.copy()

# Set red channel to zero
img_no_red[:,:,2] = 0

# Convert for display
img_no_red_rgb = cv2.cvtColor(img_no_red, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(6,6))
plt.imshow(img_no_red_rgb)
plt.title("Image with Red Channel Removed")
plt.axis("off")
plt.show()
```

```python
# Copy original image
no_blue = img.copy()

# Remove Blue channel (channel 0)
no_blue[:, :, 0] = 0

# Convert for correct display
no_blue_rgb = cv2.cvtColor(no_blue, cv2.COLOR_BGR2RGB)

# Show
plt.imshow(no_blue_rgb)
plt.title("Blue Channel Removed")
plt.axis("off")
plt.show()
```

```python
# Copy original image
no_green = img.copy()

# Remove Green channel (channel 1)
no_green[:, :, 1] = 0

# Convert for correct display
no_green_rgb = cv2.cvtColor(no_green, cv2.COLOR_BGR2RGB)

# Show
plt.imshow(no_green_rgb)
plt.title("Green Channel Removed")
plt.axis("off")
plt.show()
```
