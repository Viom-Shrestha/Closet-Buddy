```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class SimpleCNN(nn.Module):
    def __init__(self, num_classes=10):
        super(SimpleCNN, self).__init__()
        # 1st convolution: input 3 channels (RGB), output 16 channels, 3x3 kernel
        self.conv1 = nn.Conv2d(3, 16, kernel_size=3, padding=1)
        # 2nd convolution: 16 -> 32 channels
        self.conv2 = nn.Conv2d(16, 32, kernel_size=3, padding=1)
        # Max pooling layer
        self.pool = nn.MaxPool2d(2, 2)
        # Fully connected layer for classification
        self.fc1 = nn.Linear(32 * 16 * 16, num_classes)  # assuming 32x32 input images

    def forward(self, x):
        x = F.relu(self.conv1(x))
        x = self.pool(F.relu(self.conv2(x)))
        x = torch.flatten(x, 1)  # flatten all except batch
        x = self.fc1(x)
        return x

# Example usage
model = SimpleCNN(num_classes=10)
print(model)
```
